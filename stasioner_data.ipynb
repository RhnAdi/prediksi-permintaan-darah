```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for each type of stationarity

# 1. Stationary in Mean (Constant Mean)
np.random.seed(42)
time = np.arange(100)
mean_stationary = np.random.normal(loc=0, scale=1, size=100)

# 2. Stationary in Variance (Constant Variance)
variance_stationary = np.sin(2 * np.pi * time / 50) + np.random.normal(loc=0, scale=0.5, size=100)

# 3. Stationary in Covariance (Autocorrelated but Stationary)
covariance_stationary = 0.5 * np.cos(2 * np.pi * time / 50) + np.random.normal(loc=0, scale=0.5, size=100)

# Plotting
plt.figure(figsize=(15, 5))

# Plot for Stationary in Mean
plt.subplot(1, 3, 1)
plt.plot(time, mean_stationary, label="Mean Stationary")
plt.axhline(y=np.mean(mean_stationary), color='red', linestyle='--', label="Mean")
plt.title("Stationary in Mean")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# Plot for Stationary in Variance
plt.subplot(1, 3, 2)
plt.plot(time, variance_stationary, label="Variance Stationary")
plt.title("Stationary in Variance")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# Plot for Stationary in Covariance
plt.subplot(1, 3, 3)
plt.plot(time, covariance_stationary, label="Covariance Stationary")
plt.title("Stationary in Covariance")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

plt.tight_layout()
plt.show()
```
